```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
from numpy import linalg
print('Done')
```

```python
f_names=pd.read_csv('face_names.csv')
images=[]

print(plt.imread(f_names.values[0][0]))

for i in range(len(f_names)):
    images.append(plt.imread(f_names.values[i][0]))
```

```python
#Convert 2d image to 1d
def make1D(img):
    image1D=img.ravel().tolist()
    return image1D
```

```python
data=[]
for j in images:
    temp=make1D(j)
    data.append(temp)
```

```python
x=data[:len(f_names)-1]
y=data[len(f_names)-1]
```

```python
x_t=list(map(list, zip(*x)))
x_t=np.array(x_t)
xnew=list(map(list,zip(*x_t)))
xnew=np.array(xnew)
xnew
```

```python
a=np.matmul(y,x_t)
b=np.matmul(xnew,x_t)
#theta=np.matmul(linalg.inv(b),a)
theta=np.matmul(a,linalg.inv(b))
theta
```

```python
y_kappa=np.matmul(theta,xnew)
print(y[12])
print(y_kappa[12])
```

```python
finalimage = np.array(y_kappa).reshape(166,166)
targetimage = np.array(y).reshape(166,166)
```
